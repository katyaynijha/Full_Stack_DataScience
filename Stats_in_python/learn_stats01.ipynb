### STATISTICS

Two types of statistics-Descriptive and Inferential 
1. Descriptive Statistics- Organization and summarization of Data.
Types of questions descriptive stats can be use to answer-
* What is the average marks(mean,median,mode) of students in a classroom?
* What can be average weights and age of students?
* What is the relationship between weight and age of students?
NOTE- Discriptive stats widely used in EDA & FE.

2. Inferential Statistics- Technique where measurable data used to form conclusion.
Types of questions inferential stats can be use to answer-
* Are the avg age of students in a classroom less than the avg age of students in university?
* Given,50 girls' avg percentage-95% and 60 boys' avg percentage of marks-92%. Can we infer that girls have done better than boys in whole university?

**Population and samples**

Election polls of Karnataka-
All people of karnataka- Population
Some people from mysore, some people from Shimoga, some people from bangalore- Samples

In election exit polls , we want to survey which party will win. So we cannot ask each individuals, so out of whole Population , we choose Samples.
Choosing sampling require some techniques , they are-

1. Simple Random sampling- each indv has eqaul chance of being selected .
2. Stratified sampling- we divide saples in strata and sample should be taken from each strata.
e.g RBI survey on household expense- stratified
3. Systematic sampling- every nth indv will be taken for sample.
e.g sales team target nth indv for loan call or for credit card
4. Cluster sampling- divide population based on clusters and randomly select some of these cluster as samples.
5. Convenience sampling- Non probability sampling techniques where units selected where particular group is targeted.
e.g. Survey of ineuron.
6. Purposive/Judgemental sampling- members chosen based on researcher's knowledge.

### Histogram

```python
#Histograms are diffrent for discrete and continous data
# Here we are taking discrete data

from matplotlib import pyplot as plt
import numpy as np

# Creating dataset
dataset = np.random.randint(1,9,size=20)#randint- to generate random discrete data
print(dataset)

# create histogram
plt.hist(dataset,bins=10,alpha=0.5,density=True,color='r')#bin_size-assumption
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Random numbers")
```

To plot histogram for continous data we need np.random.normal(0,1,100) where 
* 1st parameter '0' is Mean- avg of normal distribution. 
* 2nd parameter '1' is standard deviation-spread of data. we take 1 to get bell-shaped curve.
* 3rd parameter- 100 random continous variable

We observe gaps in histogram with discrete data as there is no value between 30 and 40 but such gaps doesnot happen in histogram with continous data.

```python
import numpy as np
import matplotlib.pyplot as plt
# Generate dataset
cont_data=np.random.normal(0,1,100)
#print("Random Continous variables",cont_data)
# create histogram
plt.hist(cont_data,bins=10,alpha=0.5,density=True,color='r')#bin_size-assumption
plt.xlabel("Continous_Values")
plt.ylabel("Frequency")
plt.title("Histogram of continous variables")
```

### Measure of central tendency

A measure of central tendency is a single value that attempts to describe a set of data identifying the central position. 
* Mean- Average of data-(Total sum of numbers/total count)
* Median- Arrange numbers in ascending order and the middle number will give median if total count is ODD. In case of even,we will take avg of central elements(nth + (n+1)th/2).
NOTE- In case of outliers ,Median works better than Mean.3 
* Mode- It gives the frequency of number occuring often. It usually for categorical variables

```python
import numpy as np
from scipy import stats
data=np.random.randint(1,50,size=20)
print(data)
data_mean=np.mean(data)
print("Mean of data-",data_mean)
data_med=np.median(data)
print("Median of data-",data_med)
data_mod=stats.mode(data)
print("Mode of data-",data_mod)
```

### Measure of dispersion

It is the measure of the extent of variability of data about an average value. It helps to understand the distribution of the data. 

Types of dispersion- 
1. Range- Maximum value-Minimum value
2. Variance- It measures spread of data or upto how much extent the data are scattered.
Population variance-summation{(indv_data-population_mean)/N}
Sample variance- summation{(indv_data-sample_mean)/(n-1)} where (n-1) is the degree of freedom, also known as Bassel's correction.
3. Standard deviation- sqrt(variance)- Spread of data from the mean. 1st std deviation from right and left of the mean and so on .
4. Percentile and Quantiles- A percentile is a value below which a certain percentage of observation lie.
e.g. 99 percentile means the person has got better marks than 99% of  entire students.
data = [12, 34, 56, 78, 90, 45, 67, 89, 23, 45]
* Find the percentile of 67?? Let 67=x 
Following are the steps
* Arrange data in ascending order.
[12, 23, 34, 45, 45, 56, 67, 78, 89, 90]
* Apply formula, Percentile/Rank of x=((Numbers less than x)/total count)*100=6/10*100=60

Five-Number Summary-Minimum,1st quantile(Q1:25percentile),Median,3rd quantile(Q3:75 percentile),Maximum. 
We use Five-number summary to identify outliers in dataset.
* Lower fence=Q1-IQR*1.5
* Upper fence=Q3+IQR*1.5
* IQR=Q3-Q1

Note:- Numbers below than lower fence and greater than Upper fence are outliers.

```python
# Calculation of rank/percentile
import numpy as np
dataset2 = [12, 34, 56, 78, 90, 45, 67, 89, 23, 45, 44, 89, 12, 34, 49]
percentile_x=67
# calculate percentile
percentile_res=np.percentile(dataset2,percentile_x)
rank=round(percentile_res,2)
print(f"{percentile_x}th percentile: {rank}")
```
