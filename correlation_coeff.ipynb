```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import spearmanr
# calculate spearman correlation coeff between aptitude marks and resonning marks.
rm=[60,65,80,45,96,51]
am=[44,63,78,72,57,80]
# Calculate rank of both data
#rank_rm=np.argsort(rm).argsort()
#rank_am=np.argsort(am).argsort()
# Calculate spearman and pearson correlation coeff
spearman_coeff,_=spearmanr(rm,am)
pearson_coeff=np.corrcoef(rm,am)[0,1]
#print("Rank of X:", rank_rm)
#print("Rank of Y:", rank_am)
print("Spearman's Rank Correlation Coefficient:", spearman_coeff)
print("Pearson's Correlation Coefficient:", pearson_coeff)

# Plot scatter plot graph using matplotlib
plt.figure(figsize=(8,7))
plt.scatter(rm,am,label="Spearman's Ranking correlation ceff=-0.31",color='red',marker='*')
plt.title("Scatterplot")
plt.xlabel("Reasonong mrks")
plt.ylabel("Aptitude mrks")
plt.legend()
plt.grid(True)
plt.show()
```
